```python
import pandas as pd
```

```python
data = pd.read_csv("TMKOC_EPISODES_41.csv")
```

```python
def extract_runtime(rn):
    return rn.replace("FULL_HD1", "")
```

```python
data['Episode_runtime'] = data['Episode_runtime'].apply(extract_runtime)
```

```python
def extract_episode_number(s):
  parts = s.split(".")
  return int(parts[0])
```

```python
def extract_episode_title(s):
  parts = s.split(".")
  return parts[1]
```

```python
data.iloc[0:100, data.columns.get_loc('Episode_title')] = data.iloc[0:100, data.columns.get_loc('Episode_title')].apply(extract_episode_title)
```

```python
data['episode_number'] = data['episode_number'].apply(extract_episode_number)
```

```python
data['Episode_title'] = data['Episode_title'].apply(extract_episode_title)
```
